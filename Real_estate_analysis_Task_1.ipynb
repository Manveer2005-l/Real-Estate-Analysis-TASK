```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Suburb': ['Melbourne','Sydney','Brisbane','Perth'],
    'Median_Price':[850000,1200000,650000,550000],
    'Median_Income':[95000,110000,85000,80000],
    'Annual_Growth_Rate':[0.07,0.05,0.08,0.06]
}

df = pd.DataFrame(data)
df['AGR'] = (df['Median_Income']/df['Median_Price'])*df['Annual_Growth_Rate']

plt.bar(df['Suburb'],df['AGR'])
plt.title('Affordability-to-Growth Ratio (AGR) by City')
plt.xlabel('City'); plt.ylabel('AGR')
plt.show()
print(df[['Suburb','AGR']])
```
